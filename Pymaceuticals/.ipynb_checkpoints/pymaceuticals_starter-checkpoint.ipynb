```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
mouse_trial_df = pd.merge(clinical_trial_df, mouse_drug_df, on = 'Mouse ID', how = 'inner')

# Display the data table for preview
mouse_trial_df.head(10)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_volume = mouse_trial_df.groupby(['Drug','Timepoint']).mean()

# Convert to DataFrame
mean_volume = mean_volume.drop(columns =['Metastatic Sites'])

# Preview DataFrame
mean_volume.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
stderr_volume = mouse_trial_df.groupby(['Drug','Timepoint']).sem()

# Convert to DataFrame
stderr_volume = stderr_volume.drop(columns =['Metastatic Sites', 'Mouse ID'])

# Preview DataFrame
stderr_volume.head()
```

```python
#mean_volume = mean_volume.set_index('index')
#mean_volume = mean_volume.drop(columns=['level_0'])
mean_format = mean_volume.reset_index()
mean_format.head()
```

```python
reformat = mean_format.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
reformat.head()
```

```python
merged = pd.merge( mean_format, stderr_volume, on =('Drug', 'Timepoint'), how= 'inner')
merged = merged.rename(columns={'Tumor Volume (mm3)_x':'Tumor Volume (mm3)',
                               'Tumor Volume (mm3)_y':'Std_error'})
merged.head()
```

```python
capomulin = merged.loc[merged['Drug']=='Capomulin', :]
capomulin.head()

infubinol = merged.loc[merged['Drug']=='Infubinol', :]
infubinol.head()

ketapril = merged.loc[merged['Drug']=='Ketapril', :]
ketapril.head()

placebo = merged.loc[merged['Drug']=='Placebo', :]
placebo.head()
```

```python
# Generate the Plot (with Error Bars)
c1 = capomulin['Timepoint']
c2 = capomulin['Tumor Volume (mm3)']
c3 = capomulin['Std_error']

i1 = infubinol['Timepoint']
i2 = infubinol['Tumor Volume (mm3)']
i3 = infubinol['Std_error']

k1 = ketapril['Timepoint']
k2 = ketapril['Tumor Volume (mm3)']
k3 = ketapril['Std_error']

p1 = placebo['Timepoint']
p2 = placebo['Tumor Volume (mm3)']
p3 = placebo['Std_error']

fig, ax = plt.subplots()
ax.errorbar(c1, c2, c3 ,fmt="ro",linestyle='dashed', linewidth=.5, label = 'Capomulin' )
ax.errorbar(i1, i2, i3 ,fmt="b^", linestyle='dashed', linewidth=.5, label = 'Infubinol')
ax.errorbar(k1, k2, k3 ,fmt="gs", linestyle='dashed', linewidth=.5, label = 'Ketapril')
ax.errorbar(p1, p2, p3 ,fmt="d", color= 'black', linestyle='dashed', linewidth=.5, label = 'Placebo')

ax.set_xlabel("Time (days)")
ax.set_ylabel("Tumor Volume (mm3)")
ax.grid()
ax.legend()
plt.title('Tumor Response to Treatment')
# Save the Figure
plt.savefig('Tumor Response to Treatment.png')
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_meta = mouse_trial_df.groupby(['Drug','Timepoint']).mean()
# Convert to DataFrame
mean_meta = mean_meta.drop(columns=['Tumor Volume (mm3)'])

# Preview DataFrame
mean_meta.head()
```

```python
meta_reformat = mean_meta.reset_index()
meta_reformat.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
sem_meta = mouse_trial_df.groupby(['Drug','Timepoint']).sem()
# Convert to DataFrame
sem_meta = sem_meta.drop(columns = ['Mouse ID', 'Tumor Volume (mm3)'])
# Preview DataFrame
sem_meta.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat = meta_reformat.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
reformat.head()

# Preview that Reformatting worked
```

```python
merged = pd.merge( meta_reformat, sem_meta, on =('Drug', 'Timepoint'), how= 'inner')
merged = merged.rename(columns={'Metastatic Sites_x':'Metastatic Sites',
                              'Metastatic Sites_y':'Std_error'})
merged.head()
```

```python
capomulin = merged.loc[merged['Drug']=='Capomulin', :]
capomulin.head()

infubinol = merged.loc[merged['Drug']=='Infubinol', :]
infubinol.head()

ketapril = merged.loc[merged['Drug']=='Ketapril', :]
ketapril.head()

placebo = merged.loc[merged['Drug']=='Placebo', :]
placebo.head()
```

```python
# Generate the Plot (with Error Bars)
c1 = capomulin['Timepoint']
c2 = capomulin['Metastatic Sites']
c3 = capomulin['Std_error']

i1 = infubinol['Timepoint']
i2 = infubinol['Metastatic Sites']
i3 = infubinol['Std_error']

k1 = ketapril['Timepoint']
k2 = ketapril['Metastatic Sites']
k3 = ketapril['Std_error']

p1 = placebo['Timepoint']
p2 = placebo['Metastatic Sites']
p3 = placebo['Std_error']

fig, ax = plt.subplots()
ax.errorbar(c1, c2, c3 ,fmt="ro",linestyle='dashed', linewidth=.5, label = 'Capomulin' )
ax.errorbar(i1, i2, i3 ,fmt="b^", linestyle='dashed', linewidth=.5, label = 'Infubinol')
ax.errorbar(k1, k2, k3 ,fmt="gs", linestyle='dashed', linewidth=.5, label = 'Ketapril')
ax.errorbar(p1, p2, p3 ,fmt="d", color= 'black', linestyle='dashed', linewidth=.5, label = 'Placebo')

ax.set_xlabel("Treatment Duration (days)")
ax.set_ylabel("Metastatic Sites")
ax.grid()
ax.legend()
plt.title('MetaStatic Spread during Treatment')
# Save the Figure
plt.savefig('MetaStatic Spread during Treatment.png')

```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
group_df = mouse_trial_df.groupby(['Drug','Timepoint'])
# Convert to DataFrame
mouse_count = pd.DataFrame(group_df['Mouse ID'].count())
# Preview DataFrame
mouse_count.head()
```

```python
mouse_count_format = mouse_count.reset_index()
mouse_count_format.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_count_reformat = mouse_count_format.pivot(index='Timepoint',columns='Drug',values='Mouse ID')
# Preview the Data Frame
mouse_count_reformat.head()
```

```python
capomulin = mouse_count_format.loc[merged['Drug']=='Capomulin', :]
capomulin.head()

infubinol = mouse_count_format.loc[merged['Drug']=='Infubinol', :]
infubinol.head()

ketapril = mouse_count_format.loc[merged['Drug']=='Ketapril', :]
ketapril.head()

placebo = mouse_count_format.loc[merged['Drug']=='Placebo', :]
placebo.head(20)
```

```python
# Generate the Plot (Accounting for percentages)
c1 = capomulin['Timepoint']
c2 = capomulin['Mouse ID']

i1 = infubinol['Timepoint']
i2 = infubinol['Mouse ID']

k1 = ketapril['Timepoint']
k2 = ketapril['Mouse ID']

p1 = placebo['Timepoint']
p2 = placebo['Mouse ID']

#fig, ax = plt.subplots()
plt.plot(c1, c2, marker='o', color='r', linestyle='dashed', linewidth=.5, label = 'Capomulin' )
plt.plot(i1, i2, color='b', marker='^', linestyle='dashed', linewidth=.5, label = 'Infubinol')
plt.plot(k1, k2, marker='s', color = 'g', linestyle='dashed', linewidth=.5, label = 'Ketapril')
plt.plot(p1,p2, marker='o', color= 'black', linestyle='dashed', linewidth=.5, label = 'Placebo')

plt.xlabel("Treatment Duration (days)")
plt.ylabel("Survival Rate")
plt.grid()
plt.legend()
plt.title('Survival during Treatment')

# Save the Figure
plt.savefig('Survival during Treatment')

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
summary = mean_volume.reset_index()

end_values = summary.loc[summary['Timepoint'] == 45 , ['Drug', 'Tumor Volume (mm3)']]

end_values = end_values.set_index('Drug')
end_values.head()
```

```python
# Calculate the percent changes for each drug
end_values['pct_change'] = ((end_values['Tumor Volume (mm3)'] - 45) / 45)*100
end_values.head(20)
```

```python
x = np.arange(10)
y = end_values['pct_change']

mask1 = y < 0
mask2 = y > 0

# Splice the data between passing and failing drugs
fig,ax = plt.subplots()

bar1 = ax.bar(x[mask1], y[mask1], color = 'green', width= 1)
bar2 = ax.bar(x[mask2], y[mask2], color = 'red', width= 1)

# Orient widths. Add labels, tick marks, etc. 

plt.grid()
plt.ylabel('% of Total Tumor Vol Change')
plt.title('Tumor Change over 45 days')

tick_locations = [value for value in x]
ticks = summary['Drug'].unique()
plt.xticks(tick_locations, ticks, rotation=45)

plt.tight_layout()

# Use functions to label the percentages of changes
rects = ax.patches
labels = round(y)

def set_label(rects,labels):
    for rect, label in zip(rects,labels):
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width() / 2, height/2 , label,
                ha='center', va='bottom', color='w', fontweight='bold')

# Call functions to implement the function calls
set_label(rects,labels)

# Save the Figure
plt.savefig('Tumor Change over 45 Days')

# Show the Figure
fig.show()
```
