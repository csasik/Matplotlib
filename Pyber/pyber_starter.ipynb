```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
 

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

ride_df = ride_df[['city','fare','ride_id']]
```

```python
# Combine the data into a single dataset
data_df = pd.merge(ride_df, city_df,on='city', how='outer')

# Display the data table for preview
data_df.head()
```

```python
group_df = data_df.groupby('city')
group_df = group_df.agg({'fare':'mean',
                         'ride_id':'count' })
group_df = pd.merge(group_df, city_df, on = 'city', how='inner')
group_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban = group_df.loc[group_df['type']=='Urban',:]
urban.head()

x = urban['ride_id']
y=urban['fare']
z=urban['driver_count']

suburban = group_df.loc[group_df['type']=='Suburban',:]
suburban.head()

x2 = suburban['ride_id']
y2=suburban['fare']
z2=suburban['driver_count']

rural = group_df.loc[group_df['type']=='Rural',:]
rural.head()

x3 = rural['ride_id']
y3=rural['fare']
z3=rural['driver_count']

# Build the scatter plots for each city types

urban_graph = plt.scatter(x,y,s = 10*z,  linewidths=1, color = 'lightcoral', edgecolors='black', alpha=0.8 , label='Urban'     ) 

suburban_graph= plt.scatter(x2,y2,s = 10* z2, linewidths=1, color = 'lightblue', edgecolors='black', alpha=0.8 , label = 'Suburban' ) 

rural_graph= plt.scatter(x3,y3,s = 10* z3,  linewidths=1, color = 'gold', edgecolors='black', alpha=0.8 , label = 'Rural'     ) 

# Incorporate the other graph properties

plt.grid()
plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel('Total Rides Per City')
plt.ylabel('Avg Fare per City ($)')

# Create a legend
plt.legend()

# Incorporate a text label regarding circle size
plt.text(43, 38, 'Note:\nCircle size correlates with driver count per city')

# Save Figure
plt.savefig('Pyber Ride Sharing.png')
```

## Total Fares by City Type

```python
# Calculate Type Percents
total_fare = data_df['fare'].sum()

type_df = data_df.groupby('type')

type_pct =round( (type_df['fare'].sum()/total_fare)*100,1)

# Build Pie Chart
explode = (0,0,.05)
color = ['Gold','lightblue','lightcoral']
fare_pie = type_pct.plot(kind='pie', explode = explode, startangle = 140, title='% of Total Fares by City Type' ,
                     colors = color , autopct="%1.1f%%", shadow=True,    )
plt.tight_layout()


# Save Figure
plt.savefig('Total_Fares_City_Type.png')
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_rides = data_df['ride_id'].count()

ride_group = data_df.groupby('type')
ride_pct = round((ride_group['ride_id'].count()/total_rides * 100),1)

# Build Pie Chart
explode = (0,0,.1)
color = ['Gold','lightblue','lightcoral']

ride_pie = ride_pct.plot(kind='pie',  explode = explode, startangle = 140, title='% of Total Rides by City Type' ,
                     colors = color , autopct="%1.1f%%", shadow=True,    )
plt.tight_layout()


# Save Figure
plt.savefig('Total_Rides_City_Type.png')
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
total_drivers = city_df['driver_count'].sum()
print(total_drivers)

driver_group = city_df.groupby('type')
driver_pct =round( driver_group['driver_count'].sum()/total_drivers * 100, 1)
driver_pct

# Build Pie Charts
explode = (0,0,.1)
color = ['Gold','lightblue','lightcoral']

driver_pie = driver_pct.plot(kind='pie',  explode = explode, startangle = 140, title='% of Total Drivers by City Type' ,
                     colors = color , autopct="%1.1f%%", shadow=True,    )
plt.tight_layout()


# Save Figure
plt.savefig('Total_Drivers_City_Type.png')
```

```python
# Show Figure
plt.show()
```
